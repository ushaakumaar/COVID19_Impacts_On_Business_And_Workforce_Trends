```python
for key, val in state_dictionary.items():
    print(f"the key is:{key} the val is:{val}")
```

```python
# Read all the state data files into a DataFrame
for key, val in state_dictionary.items():
    file_path = os.path.join("Resources",f"{val}.csv")
    df = pd.read_csv(file_path)
    df = df.loc[(df["Year"] == 2019) | (df["Year"] == 2020)]
    df = df[["Label","Value"]]
    df = df.rename(columns={"Label" : "Month",
                       "Value" : key})
    if key == "Alabama":
        state_df = df
    else:
        state_df = pd.merge(state_df, df, on=["Month"], how="inner")
state_df
```

```python
# Set plot size
plt.figure(figsize=(15,10))

# Generate line plots for each state
for key, val in state_dictionary.items():
    plt.plot(state_df["Month"],state_df[key],marker="o",markersize=5, label=key)

# Set the orientation of the x axis labels
plt.xticks(rotation="vertical")

# Adds a legend and sets its location to best
plt.legend(loc="best")

# Set title, x and y labels 
plt.xlabel("Months")
plt.ylabel("Unemployment Rate")
plt.title("Unemployment rates for States")

# Add in a grid for the chart
plt.grid()

# Save and Display the line plot
plt.tight_layout()
plt.savefig("Images/state_lineplot.png")
plt.show()
```

```python
states_interested = ["Michigan","Hawaii","Colorado","New York", "Alaska", "Idaho","California"]

# Set plot size
plt.figure(figsize=(15,10))

# Generate line plots for each state interested
for state in states_interested:
    plt.plot(state_df["Month"],state_df[state],marker="o",markersize=5, label=state)

# Set the orientation of the x axis labels
plt.xticks(rotation="vertical")

# Adds a legend and sets its location to best
plt.legend(loc="best")

# Set title, x and y labels 
plt.xlabel("Months")
plt.ylabel("Unemployment Rate")
plt.title("Unemployment rates for States")

# Add in a grid for the chart
plt.grid()

# Save and Display the line plot
plt.tight_layout()
plt.savefig("Images/states_interested_lineplot.png")
plt.show()
```
